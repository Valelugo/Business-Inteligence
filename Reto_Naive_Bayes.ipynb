**Reto Modelo Naive Bayes** Se propone un modelo para analizar 7 variables socioeconómicas que permiten determinar si la persona vive en el área urbana o rural. Tomo como referencia la variable Urban.

0. Cargar las librerías.

```python
import numpy as np
import pandas as pd
```

1. Cargar el excel que es nuestra base de datos

```python
nxl='/content/1. BD1_Longitudinal Survey_Int.xlsx'
XDB=pd.read_excel(nxl, sheet_name=0) #La primera hoja del excel se selecciona y después se depuran los datos faltantes.
XDB.head(50)

columnas_considerar = ['Age', 'Mother_Edu', 'Father_Edu', 'FamilySize', 'Self_Esteem', 'Height', 'Weight', 'Income', 'Urban']
XDB_sin = XDB.dropna(subset=columnas_considerar)
XDB_sin.head(50)

#Se separan las variables de entrada y salida
XD=XDB_sin[['Age', 'Mother_Edu', 'Father_Edu', 'FamilySize', 'Self_Esteem', 'Height', 'Weight', 'Income']] #Variables de entrada
yd=XDB_sin[['Urban']] #Variable pronóstico

```

¿Cuál es la cantidad de datos y cuántos hay por cada variable?

```python
#Antes de eliminar las filas con datos faltantes
total_datos1 = len(XDB)
print(f"Número total de datos: {total_datos1}")

porcentaje_por_variable1 = (XDB.count() / total_datos1) * 100
print("\nPorcentaje de datos por variable:")
print(porcentaje_por_variable1)


#Después de eliminar las filas con datos faltantes
total_datos = len(XDB_sin)
print(f"Número total de datos: {total_datos}")

porcentaje_por_variable = (XDB_sin.count() / total_datos) * 100
print("\nPorcentaje de datos por variable:")
print(porcentaje_por_variable)

ND1=len(np.where((yd==1))[0])
print(ND1/len(XD)) #Viven en el área urbana

ND2=len(np.where((yd==0))[0])
print(ND2/len(XD)) #Viven en el área rural

#Se tiene en cuenta la corrección de la base de datos, de la que se eliminaron las filas que no poseían la información completa de las variables socioeconómicas seleccionadas.

```

Teniendo en cuenta que la base de datos original se modificó, ya que no eran funcionales los registros que presentaban faltantes, se evidencia un antes y después del cambio. El excel original contaba con 12.686 que, al momento de depurar, quedó con 5.948; de estos, es posible observar que el 79,16% de los ciudadanos vive en el área urbana, mientras que el 20,84% pertenece al área rural.

2. Se procede a implementar el modelo

```python
from sklearn.naive_bayes import GaussianNB

XD1=XDB_sin[['Age', 'Height', 'Weight']] #Variables de entrada
yd=XDB_sin[['Urban']] #Variable pronóstico

mnb=GaussianNB()
mnb.fit(XD,yd)


#Obtenemos la información del modelo
u=mnb.theta_ #Obtenemos la media
print("la media de las variables es: \n", u)

var=mnb.var_ #Se calculan las varianzas, me indica que tan alejados están los datos de la media
sigma=np.sqrt(var) #Se calcula la desviación
print("La desviación de las variables es: \n", sigma)

#Se determinan los límites superiores e inferiores de las tres variables
LS=u+sigma
LI=u-sigma

print("El limite superior de las variables es \n", LS)
print("El limite inferior de las variables es \n", LI)
```

Evaluar solicitante basados en sus características demográficas.

```python
#Evaluar un registro
from sklearn.naive_bayes import GaussianNB

mnb=GaussianNB() #Aloja el modelo
mnb.fit(XD,yd) #Entrena el modelo

XDS=[[34,12,15,2,24,69,160,48000]]
ydp=mnb.predict(XDS) #Se obtiene la predicción
print("La predicción del registro es:\n", ydp)

if ydp==0:
  print("El registro no vive en el area urbana")

if ydp==1:
  print("El registro vive en el area urbana")
```

Dadas las características socioeconómicas del caso hipotético, es correcto afirmar que la persona habita en el área urbana.


3. Evaluamos el desempeño del modelo

```python
from sklearn.metrics import confusion_matrix

#Haga un barrido de los datos
ydp2=mnb.predict(XD) #Haga el pronóstico para los datos base

cm=confusion_matrix(yd,ydp2)
print("La matriz de confusión es: \n", cm)
VN=cm[0,0]
FP=cm[0,1]
FN=cm[1,0]
VP=cm[1,1]

#Métricas de desempeño
Ex=(VP+VN)/(VP+VN+FP+FN) #Exactitud: Comportamiento general preaprobando
print("La exactitud del modelo es: \n", Ex)

TE=(FN+FP)/(VP+VN+FP+FN) #Tasa de error: Comportamiento general no preaprobando
print("La tasa de error del modelo es: \n", TE)

Se=VP/(VP+FN) #Sensibilidad: Comportamiento correcto preaprobando
print("La sensibilidad del modelo es: \n", Se)

Sp=VN/(VN+FP) #Especificidad: Comportamiento correcto no preaprobando
print("La especificidad del modelo es: \n", Sp)

Pr=VP/(VP+FP) #Precisión: Comportamiento correcto preaprobando
print("La precisión del modelo es: \n", Pr)

PrNeg=VN/(VN+FN) #Precisión negativa: Comportamiento correcto no preaprobando
print("La precisión negativa del modelo es: \n", PrNeg)

```

**Análisis de resultados** De acuerdo con la base de datos suministrada y depurada, la cual indica si las personas viven en el área urbana o rural, se tomaron en cuenta 8 variables socioeconómicas (edad, educación de padres (2), estatura, peso, ingresos, escala autoestima y tamaño de la familia). En esta, fue posible ubicar 5.948 registros con información completa, de los que el 80% (79,16%) viven en el área urbana y el 20% (20,84%) restante no.
En términos generales, el modelo alcanzó una exactitud superior al 75% (79,08%), lo que indica una adecuada clasificación de la variable referencia (urban), que a su vez está sustentado por una pequeña tasa de error equivalente al 20% que describe el comportamiento general para denotar el área rural. Es importante destacar que la gestión adecuada fue promovida por el índice de precisión cercano al 80% (79,18%), el cual identifica el comportamiento correcto clasificando los encuestados que habitan en la urbe.
